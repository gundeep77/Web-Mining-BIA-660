# ***Book Name: A Thousand Splendid Suns***

```python
from selenium import webdriver
import pandas as pd
import time
```

```python
def getReviews(page_url):
    global driver
    PATH = "C:\\chromedriver.exe"
    driver = webdriver.Chrome(PATH)
    driver.get(page_url)
    book_reviews_path = '//*[@id="bookReviews"]'
    final_dict = {"Name": [], "Rating": [], "Date": [], "Review": [], "Likes": []}
    
    reviews_on_first_page = len(driver.find_elements_by_class_name('friendReviews'))
    for i in range(1, 2*reviews_on_first_page, 2):
        # name
        ele_name = driver.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{1}]/span[{1}]/a')
        if ele_name.text != "":
            final_dict["Name"].append(ele_name.text)
        else:
            final_dict["Name"].append(None)

        # rating
        ele_rating = driver.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{1}]/span[{2}]/span[{1}]')
        if ele_rating.text != "":
            final_dict["Rating"].append(ele_rating.text)
        else:
            final_dict["Rating"].append(None)

        # date
        ele_date = driver.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{1}]/a')
        if ele_date.text != "":
            final_dict["Date"].append(ele_date.text)
        else:
            final_dict["Date"].append(None)

        # review
        ele_review = driver.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{2}]/span/span')
        if ele_review.text != "":
            final_dict["Review"].append(ele_review.text)
        else:
            final_dict["Review"].append(None)

        # likes
        ele_likes = driver.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{3}]/div[{1}]/span/a/span')
        if ele_likes.text != "":
            final_dict["Likes"].append(ele_likes.text.split()[0])
        else:
            final_dict["Likes"].append(None)

    reviews = pd.DataFrame(final_dict)
    return reviews
```

```python
# enter your url
page_url = "https://www.goodreads.com/book/show/128029.A_Thousand_Splendid_Suns"
getReviews(page_url)
```

```python
def getReviews_2(page_url):
    PATH = "C:\\chromedriver.exe"
    global driver2
    
    driver2 = webdriver.Chrome(executable_path=PATH)
    driver2.get(page_url)
    
    book_reviews_path = '//*[@id="bookReviews"]'
    next_page_path_1 = '//*[@id="reviews"]/div[5]/div/a[9]'
    next_page_path_2 = '//*[@id="reviews"]/div[5]/div/a[11]'

    final_dict = {"Name": [], "Rating": [], "Date": [], "Review": [], "Likes": []}
    reviews_on_first_page = len(driver2.find_elements_by_class_name('friendReviews'))
    for i in range(1, 2*reviews_on_first_page, 2):
        # name
        ele_name = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{1}]/span[{1}]/a')
        if ele_name.text != "":
            final_dict["Name"].append(ele_name.text)
        else:
            final_dict["Name"].append(None)

        # rating
        ele_rating = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{1}]/span[{2}]/span[{1}]')
        if ele_rating.text != "":
            final_dict["Rating"].append(ele_rating.text)
        else:
            final_dict["Rating"].append(None)

        # date
        ele_date = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{1}]/a')
        if ele_date.text != "":
            final_dict["Date"].append(ele_date.text)
        else:
            final_dict["Date"].append(None)

        # review
        ele_review = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{2}]/span/span')
        if ele_review.text != "":
            final_dict["Review"].append(ele_review.text)
        else:
            final_dict["Review"].append(None)

        # likes
        ele_likes = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{3}]/div[{1}]/span/a/span')
        if ele_likes.text != "":
            final_dict["Likes"].append(ele_likes.text.split()[0])
        else:
            final_dict["Likes"].append(None)
    print(driver2.current_url)

    next_page = driver2.find_element_by_xpath(next_page_path_1)
    driver2.execute_script("arguments[0].click();", next_page)
    time.sleep(2)

    for _ in range(9):
        reviews_per_page = len(driver2.find_elements_by_class_name('friendReviews'))
        for i in range(2, 2*reviews_per_page+2, 2):
            # name
            ele_name = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{1}]/span[{1}]/a')
            if ele_name.text.strip() != "":
                final_dict["Name"].append(ele_name.text)
            else:
                final_dict["Name"].append(None)

            # rating
            ele_rating = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{1}]/span[{2}]/span[{1}]')
            if ele_rating.text.strip() != "":
                final_dict["Rating"].append(ele_rating.text)
            else:
                final_dict["Rating"].append(None)

            # date
            ele_date = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{1}]/a')
            if ele_date.text.strip() != "":
                final_dict["Date"].append(ele_date.text)
            else:
                final_dict["Date"].append(None)

            # review
            ele_review = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{2}]/span/span')
            if ele_review.text.strip() != "":
                final_dict["Review"].append(ele_review.text)
            else:
                final_dict["Review"].append(None)

            # likes
            ele_likes = driver2.find_element_by_xpath(book_reviews_path + f'/div[{i}]/div/div/div/div[{3}]/div[{1}]/span/a/span')
            if ele_likes.text.strip() != "":
                final_dict["Likes"].append(ele_likes.text.split()[0])
            else:
                final_dict["Likes"].append(None)
        print(driver2.current_url)
        try: 
            next_page1 = driver2.find_element_by_xpath(next_page_path_2)
            next_page1.click()
            time.sleep(2)
            driver2.execute_script("window.scrollTo(0, 1080)")
        except: 
            break
    reviews = pd.DataFrame(final_dict)
    return reviews
```
